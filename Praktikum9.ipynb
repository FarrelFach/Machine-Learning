```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

url="/content/drive/MyDrive/Machine_Learning/petrol_consumption.csv"
dt=pd.read_csv(url)
dt
```

```python
x = dt[['Petrol_tax','Average_income','Paved_Highways','Population_Driver_licence(%)']]
y = dt['Petrol_Consumption']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
len(y_test)
```

```python
y_test
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
ccoeff_df = pd.DataFrame(regressor.coef_, x.columns, columns=['Coefficient'])
ccoeff_df
```

```python
y_pred = regressor.predict(x_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
