```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
Iris = load_iris()
dir(Iris)
```

```python
Iris.feature_names
```

```python
df = pd.DataFrame(Iris.data,columns=Iris.feature_names)
df['target']=Iris.target  #0=setosa, 1=...,2=Virginica
df['flower_name']=df.target.apply(lambda x:Iris.target_names[x])
df
```

```python
import random
random.seed(123)
ind=np.random.choice([1, 2], size=(len(df['sepal length (cm)'])), p=[0.8, 0.2])
ind
```
